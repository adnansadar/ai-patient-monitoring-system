```python
"""
Diabetes Classification Pipeline v2.0
Requirements: Python 3.9+, TensorFlow 2.15+, scikit-learn 1.4+
"""

# --------------------------
# 1. Dataset Preparation
# --------------------------
"""
Download Instructions:

A. Retinal Images (IDRiD):
1. Create IEEE account: https://ieee-dataport.org/join
2. Download dataset: https://ieee-dataport.org/open-access/indian-diabetic-retinopathy-image-dataset-idrid
3. Directory structure:
   data/
   ├── retinal/
   │   ├── train/
   │   │   ├── diabetic/
   │   │   │   ├── IDRiD_001.jpg
   │   │   │   └── ... 
   │   │   └── non_diabetic/
   │   │       ├── IDRiD_002.jpg
   │   │       └── ...
   │   └── test/
   │       ├── diabetic/
   │       └── non_diabetic/
   └── lab_reports/
       ├── diabetes.csv

B. Lab Reports (Pima Indians):
1. Download from: https://www.kaggle.com/datasets/uciml/pima-indians-diabetes-database
2. Save as 'diabetes.csv' in lab_reports folder
"""

import os
import cv2
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras import layers, optimizers, callbacks
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, roc_auc_score, ConfusionMatrixDisplay
from sklearn.utils.class_weight import compute_class_weight
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
# --------------------------
# 2. Data Loading & Verification
# --------------------------

def verify_datasets():
    """Check dataset integrity"""
    required_paths = [
        'data/retinal/train/diabetic',
        'data/retinal/train/non_diabetic',
        'data/lab_reports/diabetes.csv'
    ]
    
    missing = [p for p in required_paths if not os.path.exists(p)]
    if missing:
        raise FileNotFoundError(f"Missing datasets: {missing}")

    print("Dataset verification passed")

def load_retinal_images(img_dir, img_size=(512, 512)):
    """Load and preprocess retinal images"""
    class_names = ['non_diabetic', 'diabetic']
    images, labels = [], []
    
    for class_idx, class_name in enumerate(class_names):
        class_path = os.path.join(img_dir, class_name)
        for img_file in os.listdir(class_path):
            img_path = os.path.join(class_path, img_file)
            
            # Read and preprocess image
            img = cv2.imread(img_path)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            img = cv2.resize(img, img_size)
            
            # CLAHE contrast enhancement
            lab = cv2.cvtColor(img, cv2.COLOR_RGB2LAB)
            l_channel = lab[:,:,0]
            clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))
            cl = clahe.apply(l_channel)
            lab[:,:,0] = cl
            img = cv2.cvtColor(lab, cv2.COLOR_LAB2RGB)
            
            images.append(img.astype(np.float32)/255.0)
            labels.append(class_idx)
    
    return np.array(images), np.array(labels)

def load_lab_reports(csv_path):
    """Load and preprocess laboratory data"""
    df = pd.read_csv(csv_path)
    
    # Handle missing values (0s represent missing values in this dataset)
    df[['Glucose', 'BloodPressure', 'SkinThickness', 
        'Insulin', 'BMI']] = df[['Glucose', 'BloodPressure', 'SkinThickness',
                                'Insulin', 'BMI']].replace(0, np.nan)
    df.fillna(df.mean(), inplace=True)
    
    # Feature engineering using existing columns
    df['Glucose_BP_Ratio'] = df['Glucose'] / df['BloodPressure']
    df['Metabolic_Age'] = df['BMI'] * df['Age'] / 100
    
    # Select features
    features = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness',
                'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age',
                'Glucose_BP_Ratio', 'Metabolic_Age']
    
    return df[features], df['Outcome']


# --------------------------
# 3. Exploratory Data Analysis
# --------------------------

def perform_eda(X_img, y_img, X_lab, y_lab):
    """Comprehensive EDA for both datasets with data validation"""
    # Data validation
    assert len(X_img) == len(y_img), "Image data/label mismatch"
    assert len(X_lab) == len(y_lab), "Lab data/label mismatch"
    
    # Print class balances
    print(f"Retinal Image Class Balance:\n{pd.Series(y_img).value_counts()}")
    print(f"Lab Report Class Balance:\n{pd.Series(y_lab).value_counts()}")
    
    # Image Dataset Analysis
    diabetic_indices = np.where(y_img == 1)[0]
    non_diabetic_indices = np.where(y_img == 0)[0]
    
    plt.figure(figsize=(15, 6))
    
    plt.subplot(1, 3, 1)
    sns.countplot(x=y_img)
    plt.title('Class Distribution - Retinal Images')
    
    plt.subplot(1, 3, 2)
    plt.imshow(X_img[diabetic_indices[0]])
    plt.title(f'Diabetic Retina (Sample ID: {diabetic_indices[0]})')
    
    plt.subplot(1, 3, 3)
    plt.imshow(X_img[non_diabetic_indices[0]])
    plt.title(f'Healthy Retina (Sample ID: {non_diabetic_indices[0]})')
    
    # Lab Data Analysis
    lab_df = pd.concat([X_lab, y_lab.rename('Outcome')], axis=1)
    
    plt.figure(figsize=(18, 6))
    
    plt.subplot(1, 3, 1)
    valid_features = [col for col in ['Glucose', 'BloodPressure', 'BMI', 'Age'] if col in lab_df]
    sns.heatmap(lab_df[valid_features + ['Outcome']].corr(), annot=True, cmap='coolwarm')
    plt.title('Feature Correlation Matrix')
    
    plt.subplot(1, 3, 2)
    sns.boxplot(x='Outcome', y='BMI', data=lab_df)
    plt.title('BMI Distribution by Diabetes Status')
    
    plt.subplot(1, 3, 3)
    sns.scatterplot(data=lab_df, x='Glucose', y='Age', hue='Outcome', palette='viridis')
    plt.title('Glucose vs Age by Diabetes Status')
    
    plt.tight_layout()
    plt.show()


# --------------------------
# 4. Model Architecture & Training
# --------------------------

def build_model(input_shape, num_classes=1, learning_rate=3e-4, dropout_rate=0.5):
    """Customizable CNN architecture with hyperparameters"""
    model = tf.keras.Sequential([
        layers.Conv2D(32, (3,3), activation='relu', input_shape=input_shape),
        layers.MaxPooling2D(2,2),
        layers.BatchNormalization(),
        
        layers.Conv2D(64, (3,3), activation='relu'),
        layers.MaxPooling2D(2,2),
        layers.Dropout(dropout_rate),
        
        layers.Conv2D(128, (3,3), activation='relu'),
        layers.GlobalAveragePooling2D(),
        
        layers.Dense(256, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01)),
        layers.Dropout(dropout_rate),
        
        layers.Dense(num_classes, activation='sigmoid')
    ])
    optimizer = optimizers.AdamW(
        learning_rate=learning_rate,
        weight_decay=0.001
    )
    
    model.compile(
        optimizer=optimizer,
        loss='binary_crossentropy',
        metrics=[
            'accuracy',
            tf.keras.metrics.AUC(name='auc'),
            tf.keras.metrics.Precision(name='precision'),
            tf.keras.metrics.Recall(name='recall')
        ]
    )
    
    return model

class CustomScheduler(tf.keras.optimizers.schedules.LearningRateSchedule):
    """Custom learning rate scheduler"""
    def __init__(self, initial_learning_rate):
        self.initial_learning_rate = initial_learning_rate
        
    def __call__(self, step):
        return self.initial_learning_rate / (1 + 0.01 * step)

def train_model(model, X_train, y_train, X_val, y_val, epochs=100):
    """Training loop with advanced callbacks"""
    lr_scheduler = CustomScheduler(3e-4)
    
    # Rename the variable from 'callbacks' to 'callback_list'
    callback_list = [
        callbacks.EarlyStopping(patience=15, restore_best_weights=True),
        callbacks.ModelCheckpoint('best_model.h5', save_best_only=True),
        callbacks.LearningRateScheduler(lr_scheduler),
        callbacks.TensorBoard(log_dir='./logs')
    ]
    
    class_weights = compute_class_weight(
        class_weight='balanced',
        classes=np.unique(y_train),
        y=y_train
    )
    class_weights = dict(enumerate(class_weights))
    
    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=epochs,
        batch_size=32,
        class_weight=class_weights,
        # Update variable name here
        callbacks=callback_list,
        verbose=1
    )
    
    return history

# --------------------------
# 5. Model Evaluation & Interpretability
# --------------------------

def evaluate_model(model, X_test, y_test):
    """Comprehensive model evaluation"""
    y_pred = model.predict(X_test)
    y_pred_class = (y_pred > 0.5).astype(int)
    
    print("Classification Report:")
    print(classification_report(y_test, y_pred_class))
    
    print("\nROC AUC Score:", roc_auc_score(y_test, y_pred))
    
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 2, 1)
    ConfusionMatrixDisplay.from_predictions(y_test, y_pred_class)
    plt.title('Confusion Matrix')
    
    plt.subplot(1, 2, 2)
    fpr, tpr, _ = roc_curve(y_test, y_pred)
    plt.plot(fpr, tpr)
    plt.plot([0,1], [0,1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    
    plt.tight_layout()
    plt.show()

# --------------------------
# 6. Hyperparameter Tuning
# --------------------------

def hyperparameter_tuning(X_train, y_train):
    """Grid search for optimal hyperparameters"""
    def create_model(learning_rate=3e-4, dropout_rate=0.5):
        model = build_model(
            input_shape=X_train[0].shape,
            learning_rate=learning_rate,
            dropout_rate=dropout_rate
        )
        return model
    
    param_grid = {
        'learning_rate': [1e-4, 3e-4, 1e-3],
        'dropout_rate': [0.3, 0.5, 0.7],
        'epochs': [50, 100],
        'batch_size': [32, 64]
    }
    
    model = tf.keras.wrappers.scikit_learn.KerasClassifier(build_fn=create_model)
    grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=3)
    grid_result = grid.fit(X_train, y_train)
    
    print("Best Parameters:", grid_result.best_params_)
    return grid_result.best_estimator_

# --------------------------
# Main Execution Pipeline
# --------------------------

def main():
    # Verify and load datasets
    verify_datasets()
    
    # Load retinal images
    X_img, y_img = load_retinal_images('data/retinal/train')
    X_img_train, X_img_test, y_img_train, y_img_test = train_test_split(
        X_img, y_img, test_size=0.2, stratify=y_img, random_state=42
    )
    
    # Load lab reports
    X_lab, y_lab = load_lab_reports('data/lab_reports/diabetes.csv')
    X_lab_train, X_lab_test, y_lab_train, y_lab_test = train_test_split(
        X_lab, y_lab, test_size=0.2, stratify=y_lab, random_state=42
    )
    
    # Perform EDA
    perform_eda(X_img, y_img, X_lab, y_lab)
    
    # Build and train model
    model = build_model(X_img_train[0].shape)
    history = train_model(model, X_img_train, y_img_train, X_img_test, y_img_test)
    
    # Evaluate model
    evaluate_model(model, X_img_test, y_img_test)
    
    # Hyperparameter tuning
    best_model = hyperparameter_tuning(X_img_train, y_img_train)
    evaluate_model(best_model, X_img_test, y_img_test)

if __name__ == "__main__":
    main()
```
